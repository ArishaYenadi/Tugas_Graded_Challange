## iv. DATA CLEANING
- Membuat salinan file sumber
- Menghapus kolom yang tidak berguna
- Mengubah kolom Type 1 dan Type 2
- Mengisi missing value di kolom Type 2

```python
# Membuat file slinan agar file sumber tidak terubah
pokemon = df.copy()
```

```python
pokemon.head()
```

```python
# Menghapus kolom "#" pada tabel karena tidak diperlukan
pokemon = pokemon.drop(['#'], axis=1)
```

```python
# Mengubah kolom Type 1 menjadi Main_Type & kolom Type 2 menjadi Secondary_type
pokemon = pokemon.rename(
    columns={'Type 1': 'Primary_Type', 'Type 2': 'Secondary_Type', 'Total' : 'Total_Stats'}
)
```

```python
# Mengisi missing value pada kolom Secondary_Type yang terdapat 386 value yg kosong
pokemon['Secondary_Type'].fillna('None', inplace=True)
```

```python
# Sudah tidak ada data missing
pokemon.info()
```

```python
# Sudah tidak ada data missing
pokemon.isnull().sum()
```

```python
pokemon.head()
```

## v. EXPLORASI DATA

- Memfilter data secara spesifik
- Mengelompokan data yang dibutuhkan
- menvisualisasikan data yang dibutuhkan

```python
pokemon.head(10)
```

```python
# Mencari value Fire di dalam kolom Primary_Type (Query - 01)
Pokemon_Primary_Type_Fire = pokemon[
    pokemon['Primary_Type'] == 'Fire'
]
```

```python
Pokemon_Primary_Type_Fire.head(4)
```

```python
## Filter kolom range(index) 0-3 dan kolom ke 3 dan kolom range(index) 10-12
pokemon.iloc[:, np.r_[0:3, 3, 10:12]].head(5)
```

```python
# Filter 1 (Query - 02)
# Memfilter kolom Primary_Type, Secondary_Type, dan Sp. Atk dengan ketentuan yg diinginkan
pokemon[
    (pokemon['Primary_Type'] == 'Fire') &
    (pokemon['Secondary_Type'] == 'Dragon') &
    (pokemon['Sp. Atk'] > 90)
]
```

```python
# Filter 2 (Query - 03)
# Dan menampilkan data 8 teratas
pokemon[
    (pokemon['Primary_Type'] == 'Water') &
    (pokemon['Attack'] > 70) &
    (pokemon['Sp. Atk'] > 90)
].head(8)
```

```python
pokemon.head(4)
```

```python
# Mencari rata-rata speed dalam setiap tipe pokemon di kolom Primary_Type (Grouping - 01)
pokemon.groupby('Primary_Type')['Speed'].mean()
```

```python
# Mengkelompokan setiap tipe pokemon di dalam Primary_Type dan dicari jumlah Total_Stats di setiap tipe tersebut (Grouping - 02)
pokemon.groupby('Primary_Type')['Total_Stats'].sum()
```

```python
# Menvisualisasikan pengelompokan setiap tipe pokemon di dalam Primary_Type dan dibandingkan jumlah Total_Stats di setiap tipe tersebut
pokemon.groupby('Primary_Type')['Total_Stats'].sum().plot(kind='bar')
```

```python
# Mengkelompokan berapa banyak pokemon legendary di setiap generasinya (Grouping - 03)
pokemon.groupby('Generation')['Legendary'].sum()
```

```python
# Menvisualisasikan berapa banyak pokemon legendary di setiap generasinya
pokemon.groupby('Generation')['Legendary'].sum().plot(kind='pie')
```

```python
# Mengkelompokan pokemon tipe fire di kolom Primary_Type yang berada di generasi 1 dan 2, Dicarilah jumlah Total_Stats pokemon Fire di generasi 1 dan 2 (Grouping - 04)
pokemon[
    (pokemon['Primary_Type'] == 'Fire') &
    (pokemon['Generation'] <= 2)
].groupby(['Primary_Type', 'Generation'])['Total_Stats'].sum()
```

```python
# Menvisualisasikan berapa banyak pokemon tipe Fire di kolom Primary_Type pada generasi 1 dan 2
pokemon[
    (pokemon['Primary_Type'] == 'Fire') &
    (pokemon['Generation'] <= 2)
].groupby(['Primary_Type', 'Generation'])['Total_Stats'].sum().plot(kind='pie')
```

```python
# Menvisualisasikan Berapa banyak pokemon di berbagai tipe pokemon di kolom Primary_Type
pokemon['Primary_Type'].value_counts().plot( kind='bar')
```

```python
# Menvisualisasikan Berapa banyak pokemon di berbagai tipe pokemon di kolom Secondary_Type
pokemon['Secondary_Type'].value_counts().plot( kind='bar')
```

```python
# Menvisualisasikan berapa banyak pokemon di setiap generasinya
pokemon['Generation'].value_counts().plot( kind='pie')
```

```python
# Untuk mengubah type data object ke catagorical untuk mengurangi proses koding
pokemon['Name'] = pd.Categorical(pokemon['Name'])
```

```python
pokemon['Primary_Type'] = pd.Categorical(pokemon['Primary_Type'])
```

```python
pokemon['Secondary_Type'] = pd.Categorical(pokemon['Secondary_Type'])
```
